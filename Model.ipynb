```python
import pandas as pd
import numpy as np
```

```python
df_model = pd.read_csv('new_df_tags.csv')
```

```python
df_model.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english',binary=True)
```

```python
vector = cv.fit_transform(df_model['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend_recipes(input_ingredients_cuisine, recipes, cv):
    # Pass input ingredients as a string, not a nested list
    input_vector = cv.transform([" ".join(input_ingredients_cuisine)])
    similarities = cosine_similarity(input_vector, vector).flatten()

    # Get indices of the most similar recipes
    recommended_indices = np.argsort(similarities)[::-1]

    top_10_indices = recommended_indices[:10]
    # Access each column separately
    recommended_recipes = [{'Recipe': recipes[i]['RecipeName'],
                           'Ingredients': recipes[i]['Ingredients_name_only'],
                           'Instructions': recipes[i]['Instructions'],
                           'Time': recipes[i]['Time'],
                           'Servings': recipes[i]['Servings'],
                           'URL': recipes[i]['URL']}
                          for i in top_10_indices if similarities[i] > 0]

    return recommended_recipes
```
