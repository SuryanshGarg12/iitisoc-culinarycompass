```python
def extract_names(ingredient): #split by comma
    import re
    items = ingredient.split(',')
    names = [re.sub(r'\d+[\-/\d\s]*[a-zA-Z]*', '', item).strip() for item in items]
    return ', '.join(names)

new_df['Ingredients_name_only'] = new_df['Ingredients'].apply(extract_names)
```

```python
def extract_names(ingredient):
    items = ingredient.split(',')
    names = [item.split('-')[0].strip() for item in items]
    return ', '.join(names)

new_df['Ingredients_name_only'] = new_df['Ingredients'].apply(extract_names)
```

```python
import re
def remove_integers(ingredient):
    return re.sub(r'\d+', '', ingredient).strip()

new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_integers)
```

```python
new_df.iloc[0].Ingredients_name_only
```

```python
import re
def remove_units_and_terms(ingredient):
    units_terms = ['tablespoon', 'teaspoon', 'cup','tablespoons', 'teaspoons', 'cups','inch','grams']
    pattern = '|'.join(units_terms)
    return re.sub(pattern, '', ingredient).strip()

new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_units_and_terms)
```

```python
new_df.iloc[0].Ingredients_name_only
```

```python
def separate_parentheses(ingredient):
    return re.sub(r'\(([^)]+)\)', lambda x: ', ' + x.group(1), ingredient)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(separate_parentheses)
```

```python
new_df.iloc[0].Ingredients_name_only
```

```python
def separate_slash_terms(ingredient):
    return re.sub(r'\/\s*([^,]+)', lambda x: ', ' + x.group(1), ingredient)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(separate_slash_terms)
```

```python
def remove_pinch(ingredient):
    return ingredient.replace('inch', '').strip()
def remove_pieces(ingredient):
    return ingredient.replace('pieces', '').strip()
def remove_inches(ingredient):
    return ingredient.replace('inches', '').strip()
def remove_piece(ingredient):
    return ingredient.replace('piece', '').strip()
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_inch1)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_pieces)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_inches)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_piece)
```

```python
def remove_tsp(ingredient):
    return ingredient.replace('tsp', '').strip()
def remove_tbsp(ingredient):
    return ingredient.replace('tbsp', '').strip()
def remove_tsps(ingredient):
    return ingredient.replace('tsps', '').strip()
def remove_tbsps(ingredient):
    return ingredient.replace('tbsps', '').strip()
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_tsp)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_tbsp)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_tsps)
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_tbsps)
```

```python
def remove_sprig(ingredient):
    return ingredient.replace('sprig', '').strip()
new_df['Ingredients_name_only'] = new_df['Ingredients_name_only'].apply(remove_sprig)
```
